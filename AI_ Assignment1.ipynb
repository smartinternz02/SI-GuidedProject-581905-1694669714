```python
import numpy as np
import pandas as pd

#TASK 1

rows = 4
cols = 4
random_values = np.random.rand(rows, cols)

df = pd.DataFrame(random_values, columns = ["Column1","Column2","Column3","Column4"])
print(df)
```

```python
#Task 2
new_names = {
    "Column1" : "Random value 1",
    "Column2" : "Random value 2",
    "Column3" : "Random value 3",
    "Column4" : "Random value 4"
}

df.rename(columns = new_names, inplace = True)
print(df)
```

```python
#Task 3

df.describe()
```

```python
#Task 4

print("Checking for null values \n", df.isnull())
print("\n\n Showing data types \n", df.dtypes)
```

```python
#Task 5

print("Printing by location method")
row = df.loc[:,'Random value 2' : 'Random value 3']
print(row)

print("\nPrinting by index location method")
val = df.iloc[:,1:3]
print(val)
```
