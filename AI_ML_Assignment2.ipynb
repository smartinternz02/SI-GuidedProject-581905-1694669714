```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/House Price India.csv")
dataset.head()
```

#Univariate Analysis

```python
uni_var = 'Price'
print(dataset[uni_var].describe())
```

##Histrogram

```python
plt.figure(figsize=(8,5))
plt.hist(dataset[uni_var], bins =20, color = "orange", edgecolor = "black")
plt.title(f'Distribution of {uni_var}')
plt.xlabel(uni_var)
plt.ylabel('Frequency')
plt.show()
```

##Histogram using seaborn

```python
plt.figure(figsize=(8,5))
sns.histplot(data=dataset, x =uni_var, bins = 20, kde=True, color='orange')
plt.title(f'Distribution of {uni_var}')
plt.xlabel(uni_var)
plt.ylabel('Frequency')
plt.show()
```

##Boxplot


```python
plt.figure(figsize = (8,5))
sns.boxplot(data=dataset, x=uni_var, color = 'orange')
plt.title(f'Box plot of {uni_var} ')
plt.xlabel(uni_var)
plt.show()
```

##Bivaraite Analysis

```python
var_x = 'number of bedrooms'
var_y = 'living area'
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=dataset, x=var_x, y=var_y,marker= 'o', color = 'blue')
plt.title('Bivariate Line plot with Seaborn')
plt.xlabel(var_x)
plt.ylabel(var_y)
plt.grid(True)
plt.show()
```

##Scatter plot

```python
plt.figure(figsize=(8,5))
plt.scatter(dataset[var_x], dataset[var_y], color = 'orange', alpha=1)
plt.title(f'{var_x} vs {var_y}')
plt.xlabel(var_x)
plt.ylabel(var_y)
plt.show()
```

##Scatter using seaborn

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=dataset, x=var_x, y=var_y, color = 'orange', alpha=1)
plt.title(f'{var_x} vs {var_y}')
plt.xlabel(var_x)
plt.ylabel(var_y)
plt.show()
```

#Multivariate
##Heat map

```python
var_z='Price'
var_corr= [var_x,var_y,var_z]
correlation_matrix = dataset[var_corr].corr()

plt.figure(figsize=(8,5))
sns.heatmap(correlation_matrix, annot = True, cmap = "coolwarm", linewidth = 1)
plt.title(f'Correlation Heatmap between {",".join(var_corr)}')
plt.show()
```

##Pair plot

```python
sns.pairplot(data=dataset, vars = [var_x,var_y,var_z], diag_kind= 'kde', kind='scatter')
plt.suptitle(f'Pair Plot of {var_x, var_y,var_z}')
plt.show()
```
